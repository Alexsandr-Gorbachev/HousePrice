```python
CSV = 'cianhouse11.05_1part_in2_5.csv'
```

```python
HOST = 'https://nn.cian.ru/'
```

```python
URL = 'https://nn.cian.ru/cat.php?currency=2&deal_type=sale&engine_version=2&maxprice=5000000&minprice=2000000&object_type%5B0%5D=1&offer_type=suburban&p&region=4596'
```

```python
HEADERS = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36 Edg/100.0.1185.44'
}
```

```python
def get_html(url):
    r = requests.get(url, headers=HEADERS)
    return r

# html = get_html(URL)

def get_content(html):
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_='_93444fe79c--card--ibP42')
    cards = []
#     print(items)
    
# html = get_html(URL)
# get_content(html.text)

    for item in items:
        time.sleep(1)
        cards.append(
            {
                'general':item.find('div', class_='_93444fe79c--general--BCXJ4').get_text(strip=True),
                'time_lab':item.find('div', class_='_93444fe79c--timeLabel--_VT26').get_text(strip=True),
                'price_out':item.find('div', class_='_93444fe79c--container--aWzpE').get_text(strip=True),
                'link_product':item.find('div', class_='_93444fe79c--general--BCXJ4').find('a').get('href')
            }
        )
        
    
    return cards


def save_doc(items, path):
    with open(path, 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file, delimiter=';')
        writer.writerow(['Главное', 'Время_разм', 'Цена_внеш', 'Ссылка'])
        for item in items:
            writer.writerow( [item['general'], item['time_lab'], item['price_out'], item['link_product']] )
            

def parser():
    PAGENATION = input('Укажите количество страниц для парсинга: ')
    PAGENATION = int(PAGENATION.strip())
    html = get_html(URL)
    if html.status_code==200:
        cards = []
        for page in range(1, PAGENATION):
            time.sleep(1)
            print(f'Парсим страницу: {page}')
            html = get_html(f'https://nn.cian.ru/cat.php?currency=2&deal_type=sale&engine_version=2&maxprice=5000000&minprice=2000000&object_type%5B0%5D=1&offer_type=suburban&p={page}&region=4596')
            cards.extend(get_content(html.text))
            save_doc(cards, CSV)
        pass
    else:
        print('Error')
        
parser()
```

```python
link = in_cart['Ссылка'][:500]
```

```python
user_database = []
for user in tqdm(link):
    time.sleep(randrange(6, 12))
    r = requests.get(user, headers=HEADERS)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    try:
        tot_area = soup.find('div', {'class': 'a10a3f92e9--container--TrMPA'}).text
    except:
        tot_area = "Нет"
        
        time.sleep(randrange(1, 3))
    try:
        desc_parts = soup.find('div', {'class': 'a10a3f92e9--description-parts--O2ZC4'}).text
    except:
        desc_parts = "Нет"
        
    try:
        sel_div = soup.find('div', {'class': 'a10a3f92e9--section_divider--Ox1l7'}).text
    except:
        sel_div = "Нет"
        
    try:
        geo = soup.find('div', {'class': 'a10a3f92e9--geo--VTC9X'}).text
    except:
        geo = "Нет"
        
    try:
        time.sleep(randrange(1, 3))
        price_per = soup.find('div', {'class': 'a10a3f92e9--price_per_meter--yfcbi a10a3f92e9--price_per_meter--residential--DUaaw'}).text
    except:
        price_per = "Нет"
        
    try:
        price = soup.find('span', {'class': 'a10a3f92e9--price_value--lqIK0'}).text
    except:
        price = "Нет"
        
    try:
        views = soup.find('div', {'class': 'a10a3f92e9--container--L22PY'}).text    
    except:
        views = "Нет"
        
    try:
        time.sleep(randrange(1, 3))
        rate = soup.find('div', {'class': 'a10a3f92e9--container--hC7C1'}).text
    except:
        rate = "Нет"
#     face = soup.find('div', {'class': 'a10a3f92e9--container--KfHco'}).text

    user_data = {
        'tot_area': tot_area, 'desc_parts': desc_parts, 'sel_div': sel_div, 'geo': geo, 'price_per': price_per, 
        'price': price, 'views': views, 'rate': rate
    }
    user_database.append(user_data)
```

```python
def save_doc(items):
    with open('cianhouse11.05_1part_out2_5', 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file, delimiter=';')
        writer.writerow(['Общая пл', 'Описание', 'Общ инф', 'Адрес', 'Цена за М2', 'Цена', 'Просмотры', 'Тариф'])
        for item in items:
            writer.writerow( [item['tot_area'], item['desc_parts'], item['sel_div'], item['geo'], item['price_per'], 
                              item['price'], item['views'], item['rate']] )
            
save_doc(user_database)
```
